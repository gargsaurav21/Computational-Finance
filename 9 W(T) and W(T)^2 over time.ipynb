$$
\textbf{\large Summary: Evolution of } W(T) \textbf{ and } W(T)^2 \textbf{ Over Time}
$$

$$
\textbf{Brownian Motion:} \quad W(T) \sim N(0, T)
$$

$$
\begin{aligned}
&\text{Brownian motion } W(T) \text{ follows a normal distribution:} \\
&\quad W(T) \sim N(0, T), \text{ where mean } E[W(T)] = 0 \text{ and variance } \text{Var}(W(T)) = T. \\
&\text{At time } T, \text{ the standard deviation grows as } \sqrt{T}, \text{ causing greater spread.}
\end{aligned}
$$

$$
\textbf{Squared Process:} \quad W(T)^2 \sim T \cdot \chi^2_1
$$

$$
\begin{aligned}
&\text{Since } W(T) \text{ is normally distributed, its squared process follows a scaled chi-squared distribution:} \\
&\quad W(T)^2 \sim T \cdot \chi^2_1. \\
&\text{Key properties:} \\
&\quad E[W(T)^2] = T, \quad \text{Var}(W(T)^2) = 2T^2. \\
&\quad \text{It is positively skewed, meaning a longer tail on the right.}
\end{aligned}
$$

$$
\textbf{Visualization Results:}
$$

$$
\begin{array}{|c|c|c|}
\hline
T & W(T) \sim N(0, T) & W(T)^2 \sim T \cdot \chi^2_1 \\
\hline
1 & \text{N(0,1): Symmetric normal} & \text{Chi-squared: Skewed right} \\
\hline
2 & \text{N(0,2): Wider spread} & \text{Chi-squared: Mean = 2, larger spread} \\
\hline
5 & \text{N(0,5): Even wider} & \text{Chi-squared: Mean = 5, variance growing} \\
\hline
10 & \text{N(0,10): Largest spread} & \text{Chi-squared: Strong right-skew} \\
\hline
\end{array}
$$

$$
\text{A Python simulation was provided to visualize these distributions at different time intervals.} \\
\text{The results confirm that as } T \text{ increases, } W(T) \text{ spreads symmetrically, while } W(T)^2 \text{ shifts rightward.}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters
num_samples = 10000  # Number of Brownian paths
time_intervals = [1, 2, 5, 10]  # Different time points

fig, axes = plt.subplots(2, len(time_intervals), figsize=(14, 6))

for i, T in enumerate(time_intervals):
    # Generate samples
    W_samples = np.random.normal(0, np.sqrt(T), num_samples)  # W(T) ~ N(0, T)
    W_squared_samples = W_samples ** 2  # W(T)^2 follows a scaled chi-squared distribution
    
    # Plot W(T)
    sns.histplot(W_samples, bins=50, kde=True, ax=axes[0, i], color="blue")
    axes[0, i].set_title(f"W(T) at T={T}")
    axes[0, i].set_xlabel("Value")
    
    # Plot W(T)^2
    sns.histplot(W_squared_samples, bins=50, kde=True, ax=axes[1, i], color="red")
    axes[1, i].set_title(f"W(T)^2 at T={T}")
    axes[1, i].set_xlabel("Value")

# Formatting
axes[0, 0].set_ylabel("Density (W(T))")
axes[1, 0].set_ylabel("Density (W(T)^2)")
plt.suptitle("Comparison of W(T) and W(T)^2 Over Time", fontsize=14)
plt.tight_layout()
plt.show()
```
