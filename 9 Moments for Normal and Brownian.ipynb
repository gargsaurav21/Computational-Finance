$$
\textbf{\large Summary:} \\ \\
\text{We compare the first four moments (raw and central) for two distributions:} \\ \\
\quad (i) \text{Normal Distribution: } X \sim N(0,\sigma^2) \\ \\
\quad (ii) \text{Brownian Motion: } W(t) \sim N(0,t) \quad \text{(at time } t\text{)} \\ \\
\textbf{Key Observations:} \\ \\
\quad \bullet \text{Both distributions are symmetric, so the 1st and 3rd moments are zero.} \\ \\
\quad \bullet \text{The 2nd moment represents variance: } \sigma^2 \text{ for } X \text{ and } t \text{ for } W(t). \\ \\
\quad \bullet \text{The 4th raw moment is } 3\sigma^4 \text{ for } X \text{ and } 3t^2 \text{ for } W(t). \\ \\
\quad \bullet \text{Since the mean is } 0, \text{ the raw and central moments coincide.}
$$

$$
\textbf{Raw Moments:}
$$

$$
\begin{array}{|c|c|c|}
\hline
\textbf{Moment} & \textbf{Normal Distribution } (X \sim N(0,\sigma^2)) & \textbf{Brownian Motion } (W(t) \sim N(0,t)) \\
\hline
\text{1st Moment} & E[X] = 0 & E[W(t)] = 0 \\
\hline
\text{2nd Moment} & E[X^2] = \sigma^2 & E[W(t)^2] = t \\
\hline
\text{3rd Moment} & E[X^3] = 0 & E[W(t)^3] = 0 \\
\hline
\text{4th Moment} & E[X^4] = 3\sigma^4 & E[W(t)^4] = 3t^2 \\
\hline
\end{array}
$$

$$
\textbf{Central Moments:}
$$

$$
\begin{array}{|c|c|c|}
\hline
\textbf{Central Moment} & \textbf{Normal Distribution } (X \sim N(0,\sigma^2)) & \textbf{Brownian Motion } (W(t) \sim N(0,t)) \\
\hline
\text{1st Central Moment} & E[X - E[X]] = 0 & E[W(t) - 0] = 0 \\
\hline
\text{2nd Central Moment} & E[(X - E[X])^2] = \sigma^2 & E[(W(t) - 0)^2] = t \\
\hline
\text{3rd Central Moment} & E[(X - E[X])^3] = 0 & E[(W(t) - 0)^3] = 0 \\
\hline
\text{4th Central Moment} & E[(X - E[X])^4] = 3\sigma^4 & E[(W(t) - 0)^4] = 3t^2 \\
\hline
\end{array}
$$

$$
\text{A Python simulation was also provided to compute moments and visualize distributions.} \\ \\
\text{For } t = 1, \text{ Brownian motion } W(1) \text{ is equivalent to } N(0,1).
$$


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters
T = 1.0  # Time horizon for Brownian motion (variance t for Brownian motion at time t)
N = 10000  # Number of samples

# Generate samples from a standard normal distribution: N(0,1)
normal_samples = np.random.normal(0, 1, N)

# For Brownian motion at time T, the distribution is N(0, T) hence std = sqrt(T)
brownian_samples = np.random.normal(0, np.sqrt(T), N)

# Function to compute central moments: Mean, Variance, Skewness, and Kurtosis
def compute_central_moments(samples):
    mean = np.mean(samples)
    variance = np.var(samples)
    skewness = stats.skew(samples)
    kurtosis = stats.kurtosis(samples, fisher=False)  # fisher=False returns the "raw" kurtosis (normal -> 3)
    return mean, variance, skewness, kurtosis

# Compute the moments for each distribution
normal_moments = compute_central_moments(normal_samples)
brownian_moments = compute_central_moments(brownian_samples)

# Print results
print("Normal Distribution Central Moments:")
print(f"Mean: {normal_moments[0]:.4f}, Variance: {normal_moments[1]:.4f}, "
      f"Skewness: {normal_moments[2]:.4f}, Kurtosis: {normal_moments[3]:.4f}")

print("\nBrownian Motion (t = 1) Central Moments:")
print(f"Mean: {brownian_moments[0]:.4f}, Variance: {brownian_moments[1]:.4f}, "
      f"Skewness: {brownian_moments[2]:.4f}, Kurtosis: {brownian_moments[3]:.4f}")

# Visualization: plot histograms to illustrate the distributions
plt.figure(figsize=(10, 5))
plt.hist(normal_samples, bins=50, alpha=0.6, density=True, label="Normal Distribution (N(0,1))")
plt.hist(brownian_samples, bins=50, alpha=0.6, density=True, label="Brownian Motion (W(1) ~ N(0,1))")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram: Normal Distribution vs. Brownian Motion")
plt.legend()
plt.show()
```
