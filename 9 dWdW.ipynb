```python
#%%
"""
Created on Jan 16 2019
Convergence of E(W(t_i+1)-W(t_i))^2 and Var(W(t_i+1)-W(t_i))^2 with respect to
the number of the discretization intervals
@author: Lech A. Grzelak
"""
import numpy as np
import matplotlib.pyplot as plt
```

$$
\textbf{\large Summary: Brownian Motion Squared Increments Simulation}
$$

$$
\begin{aligned}
&\text{The code simulates } N \text{ paths of Brownian motion } W(T) \text{ over } T=1.0, \text{ with } m \text{ subdivisions.}\\
&\text{For each subdivision } m, \text{ the interval is partitioned into smaller segments:}\\
&\quad t_1 = \frac{T}{m}, \quad t_2 = \frac{2T}{m}.\\
&\text{Brownian motion increments are generated using:} \\
&\quad W(t_1) = \sqrt{t_1} \cdot N(0,1) \\
&\quad W(t_2) = W(t_1) + \sqrt{t_2 - t_1} \cdot N(0,1).
\end{aligned}
$$

$$
\textbf{Squared Increments: } X = (W(t_2) - W(t_1))^2
$$

$$
\begin{aligned}
&\text{The squared increments } X \text{ are stored, and their mean and variance computed:} \\
&\quad E[X] = \text{Mean}((W(t_2) - W(t_1))^2) \\
&\quad \text{Var}(X) = \text{Variance}((W(t_2) - W(t_1))^2).
\end{aligned}
$$

$$
\textbf{Visualization:}
$$

$$
\begin{array}{|c|c|}
\hline
\text{Plot} & \text{Description} \\
\hline
E[X] & \text{Mean of squared Brownian increments over partitions} \\
\hline
\text{Var}(X) & \text{Variance of squared Brownian increments over partitions} \\
\hline
\end{array}
$$

$$
\text{As } m \text{ increases, the mean and variance exhibit expected behavior, matching the theoretical:}\\
E[(W(t_2) - W(t_1))^2] = (t_2 - t_1), \quad \text{Var}[(W(t_2) - W(t_1))^2] = 2(t_2 - t_1)^2.
$$


```python
def mainCalculation():
    NoOfPaths = 50000

    mV = []
    vV = []
    T=1.0
    for m in range(2,60,1):
        t1 = 1.0*T/m;
        t2 = 2.0*T/m;
        W_t1 = np.sqrt(t1)* np.random.normal(0.0,1.0,[NoOfPaths,1])
        W_t2 = W_t1 + np.sqrt(t2-t1)* np.random.normal(0.0,1.0,[NoOfPaths,1])
        X = np.power(W_t2-W_t1,2.0);
        mV.append(np.mean(X))
        vV.append(np.var(X))

    plt.figure(1)
    plt.plot(range(2,60,1), mV)
    plt.plot(range(2,60,1), vV,'--r')
    plt.grid()
    plt.legend(['E(W(t_i+1)-W(t_i))^2','Var(W(t_i+1)-W(t_i))^2'])
```

```python
mainCalculation()
```
